Предположим вы получили на контесте задачу, в которой необходимо сделать нечто похожее: найти самый длинный(короткий) хороший отрезок или посчитать количество хороших отрезков.
Как понять, можно ли применить к ней метод двух указателей?

Во первых, должно выполняться одно из двух следующих свойств:

    если отрезок [L,R]
    хороший, то любой вложенный в него отрезок также хороший (в таком случае можно применить код из первой задачи);

    если отрезок [L,R]
    хороший, то любой отрезок, который его содержит также хороший (в таком случае можно применить код из второй задачи).
Во вторых, вы должны уметь пересчитывать вашу функцию (проверять хороший отрезок или плохой), при увеличении левой или правой границы на единицу вправо.

В подобных задачах, код почти всегда выглядит так:

L = 0
for R = 0..n-1

    add(a[R])
    while not good():
          remove(a[L])
        L++
Для решения подобных задач нужно реализовать три функции: add, remove и good.
